```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
```

```python
sns.set_style('white')
```

```python
path = os.path.join('data')
```

## Heart disease dataset

```python
cols = [f'X{i+1}' for i in range(9)]
cols = cols + ['TARGET']
data = pd.DataFrame(np.genfromtxt(os.path.join('data','DATA.txt')), columns=cols)
```

```python
data.describe()
```

```python
ax = sns.histplot(data, x='TARGET',  stat='percent') 
for p in ax.patches:
    if p.get_height() != 0.0:
        ax.annotate(f"{format(p.get_height(), '.1f')} %", (p.get_x() + p.get_width() / 2., p.get_height()), ha = 'center', va = 'center', xytext = (0, 5), textcoords = 'offset points')
plt.title('Distribution of the classes - Heart disease dataset', fontdict=dict(size=18))
plt.xlabel('Class', fontdict=dict(size=15))
plt.ylabel('Percentage', fontdict=dict(size=15))
plt.savefig('heart-hist.png')
```
